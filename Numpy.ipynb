```python
# 42. 다음 ndarray 객체에서 음수를 0으로 치환하세요
arr = np.array([-1, 2, -3, 4])
arr[arr<0] = 0
arr
```

```python
# 43. 다음 ndarray 객체에서 음수를 양수로 변환하세요
arr = np.array([-1, 2, -3, 4])
arr[arr<0] = -arr[arr<0]
arr
```

```python
# 44. 다음 ndarray 객체에서 양수 혹은 0이면 10을 더하고 음수는 10을 빼세요.
arr = np.array([-1, 2, -3, 4])
arr[arr>=0] = arr[arr>=0] + 10
arr[arr<0] = arr[arr<0] - 10
arr
```

```python
# 45. numpy random 모듈의 shuffle 함수를 사용해서 입력 리스트의 원소들을 랜덤하게 섞어 보세요.
arr = [1, 2, 3, 4, 5]
np.random.shuffle(arr)
```

```python
# 48.다음 넘파이 배열을 오름차순 정렬해보세요.
arr = np.array([2, 1, 5, 3, 7, 4, 6, 8])
np.sort(arr)
```

```python
# 49. numpy.vstack 함수를 사용하여 다음 2차원 배열을 생성하세요.
arr1 = np.array([1, 2])
arr2 = np.array([3, 4])

np.vstack([arr1,arr2]) # 행 쌓기
np.hstack([arr1,arr2]) # 열 쌓기
```
